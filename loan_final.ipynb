```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline , make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
#
dataset=pd.read_csv('train_u6lujuX_CVtuZ9i (1).csv')
```

```python
dataset
```

```python
dataset['Dependents']=dataset['Dependents'].replace({'3+':3} ).astype(float)
```

```python
dataset
```

```python
dataset['Loan_Status']=dataset['Loan_Status'].replace({'N':0 , 'Y':1})
```

```python
dataset.columns
```

```python
X=dataset.drop(columns=['Loan_ID' ,'Loan_Status'])
```

```python
X
```

```python
y=dataset['Loan_Status']
```

```python
y
```

```python
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
X_train
```

```python
num_col=X_train[['Dependents', 'ApplicantIncome' ,'CoapplicantIncome','LoanAmount']].columns
```

```python
num_col
```

```python
cat_num=X_train[['Loan_Amount_Term','Credit_History']].columns
```

```python
cat_num
```

```python
cat_ohe=X_train[['Gender' ,'Married' ,'Education', 'Self_Employed','Property_Area']].columns
```

```python
cat_ohe
```

```python
pipe1=Pipeline([
    ('p1', SimpleImputer(strategy='mean')),
    ('sc' ,RobustScaler())
])
```

```python
pip2=Pipeline([
    ('p2' ,SimpleImputer(strategy='most_frequent') ),
    ('scs' , RobustScaler())
])
```

```python
pipe3=Pipeline([
    ('sd' , SimpleImputer(strategy='most_frequent')),
    ('ohe' , OneHotEncoder(sparse=False , drop='first'))
])
```

```python
t1=ColumnTransformer([
    ('t1' ,pipe1 , num_col),
    ('t2' ,pip2 , cat_num),
    ('t3' ,pipe3 , cat_ohe)
])
```

```python
t2=LogisticRegression()
```

```python
pipe4=Pipeline([
    ('z1' , t1),
    ('z2' ,t2)
])
```

```python
pipe4.fit(X_train ,y_train)
```

```python
y_pred=pipe4.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test ,y_pred)
```

```python
pipe4.score
```
